```python
def alterar_colunas_para_snake_case(df, ignorar=None, renomear_ignorar=None):
    if ignorar is None:
        ignorar = []
    if renomear_ignorar is None:
        renomear_ignorar = {}

    # Função auxiliar para converter para snake_case
    def para_snake_case(nome):
        # Adiciona underscore antes das letras maiúsculas e transforma tudo em minúsculo
        nome_snake = re.sub(r'(?<!^)(?=[A-Z])', '_', nome).lower()
        return nome_snake

    # Renomeia as colunas, exceto as especificadas em ignorar
    novas_colunas = {
        col: para_snake_case(col) if col not in ignorar else col
        for col in df.columns
    }
    df.rename(columns=novas_colunas, inplace=True)

    # Renomeia as colunas ignoradas manualmente, conforme especificado
    df.rename(columns=renomear_ignorar, inplace=True)

    return df

# Colunas a ignorar
colunas_ignorar = ['PrecoIPI', 'PrecoICMS', 'P_IPI', 'P_ICM', 'UF', 'Data_Pedido', 'Data_Nota']

# Novos nomes para as colunas ignoradas
renomear_ignorar = {
    'PrecoIPI': 'preco_ipi',
    'PrecoICMS': 'preco_icms',
    'P_IPI': 'p_ipi',
    'P_ICM': 'p_icm',
    'UF': 'uf',
    'Data_Pedido': 'data_pedido',
    'Data_Nota': 'data_nota'
}

# Aplicando a função
df = alterar_colunas_para_snake_case(df, ignorar=colunas_ignorar, renomear_ignorar=renomear_ignorar)
```

```python
#problemas com os datatypes
# Converter as colunas para os tipos apropriados
df['data_nota'] = pd.to_datetime(df['data_nota'])    # Converter para datetime
df['cd_cliente'] = df['cd_cliente'].astype(str)      # Converter para object
df['cd_operacao'] = df['cd_operacao'].astype(str)    # Converter para object
df['nota'] = df['nota'].astype(str)                  # Converter para object
df['cd_produto'] = df['cd_produto'].astype('str')    # Converter para object
```
