# Resolução Atividade - Capítulo Matplotlib

## 1 - Módulos e dados

- Módulos

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

- Dados

```python
dados = pd.read_csv('dados_1997_2011_paises_csv.csv',
                    sep = ';',
                    decimal = ',',
                    encoding = 'latin1')
```

## Questão 2

Gerar um gráfico de linhas que mostra a evolução do IDH do Brasil durante os anos.

```python
dados_brasil = dados[dados['pais'] == 'Brasil']
```

```python
plt.plot(dados_brasil['ano'], dados_brasil['idh'])
```

## Questão 3

Gerar um gráfico de barras para comparar o Brasil e E.U.A no ano de 2009 em relação ao índice de globalização.

```python
dados_filtrados = dados[(dados['pais'].isin(['Brasil', 'Estados Unidos'])) & \
                        (dados['ano'] == 2009)]
```

```python
plt.bar(dados_filtrados['pais'], dados_filtrados['globalizacao_indice'])
```
