## 1 - Módulos e dados

- Módulos

```python
import numpy as np
import pandas as pd
```

- Dados

```python
dados = pd.read_csv('dados_1997_2011_paises_csv.csv', 
                    sep = ';', 
                    decimal = ',', 
                    encoding = 'latin1')
```

## 2 - Filtrar e analisar Brasil

Filtrar o país Brasil e obtenha as medidas: média, mediana, mínimo, máximo para as colunas “idh” e “competitividade_indice”.

```python
dados_brasil = dados[dados['pais'] == 'Brasil'].reset_index(drop=True)
```

```python
# idh
dados_brasil.agg(min_idh = pd.NamedAgg('idh', 'min'),
                 max_idh = pd.NamedAgg('idh', 'max'),
                 media_idh = pd.NamedAgg('idh', 'mean'),
                 mediana_idh = pd.NamedAgg('idh', 'median'))
```

```python
# competitividade_indice
dados_brasil.agg(min_compet = pd.NamedAgg('competitividade_indice', 'min'),
                 max_compet = pd.NamedAgg('competitividade_indice', 'max'),
                 media_compet = pd.NamedAgg('competitividade_indice', 'mean'),
                 mediana_compet = pd.NamedAgg('competitividade_indice', 'median'))
```

## Filtrar e analisar Alemanha

Filtrar o país Alemanha e obter as mesmas medidas analisadas no Brasil. Discorra sobre quais foram as principais diferenças encontradas.

```python
dados_alemanha = dados[dados['pais'] == 'Alemanha'].reset_index(drop=True)
```

```python
# idh
dados_alemanha.agg(min_idh = pd.NamedAgg('idh', 'min'),
                 max_idh = pd.NamedAgg('idh', 'max'),
                 media_idh = pd.NamedAgg('idh', 'mean'),
                 mediana_idh = pd.NamedAgg('idh', 'median'))
```

```python
# competitividade_indice
dados_alemanha.agg(min_compet = pd.NamedAgg('competitividade_indice', 'min'),
                 max_compet = pd.NamedAgg('competitividade_indice', 'max'),
                 media_compet = pd.NamedAgg('competitividade_indice', 'mean'),
                 mediana_compet = pd.NamedAgg('competitividade_indice', 'median'))
```

- Comentários

A Alemanha apresenta índices bem mais elevados que o Brasil em relação ao IDH e Competitividade. O Brasil apresenta evolução, porém ainda possui um longo caminho até obter os níveis da Alemanha.
